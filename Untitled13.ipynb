```python
test = data.dropna()
```

```python
test.isnull().sum()
```

```python
test.columns
```

```python
gender_satisfaction_counts = test.groupby(['Gender', 'satisfaction']).size().reset_index(name='count')
print(gender_satisfaction_counts)
```

```python
customer_type_satisfaction_counts = test.groupby(['Customer Type', 'satisfaction']).size().reset_index(name='count')
print(customer_type_satisfaction_counts)
```

```python
class_satisfaction_counts = test.groupby(['Class', 'satisfaction']).size().reset_index(name='count')
print("\nClass vs Satisfaction (Counts):")
print(class_satisfaction_counts)
```

```python
food_drink_satisfaction_counts = test.groupby(['Food and drink', 'satisfaction']).size().reset_index(name='count')
print("\nFood and Drink vs Satisfaction (Counts):")
print(food_drink_satisfaction_counts)
```

```python
seat_comfort_satisfaction_counts = test.groupby(['Seat comfort', 'satisfaction']).size().reset_index(name='count')
print("\nSeat Comfort vs Satisfaction (Counts):")
print(seat_comfort_satisfaction_counts)
```

```python
sns.countplot(data=test, x='Cleanliness', hue='satisfaction')
plt.title('Cleanliness vs Satisfaction')
plt.show()
```

```python
sns.countplot(data=test, x='Inflight wifi service', hue='satisfaction')
plt.title('Inflight Wifi Service vs Satisfaction')
plt.xlabel('Inflight Wifi Service Rating')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(data=test, x='Baggage handling', hue='satisfaction')
plt.title('Baggage Handling vs Satisfaction')
plt.xlabel('Baggage Handling Rating')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(data=test, x='Ease of Online booking', hue='satisfaction')
plt.title('Ease of Online Booking vs Satisfaction')
plt.xlabel('Ease of Online Booking Rating')
plt.ylabel('Count')
plt.show()
```

```python
leg_room_satisfaction_counts = test.groupby(['Leg room service', 'satisfaction']).size().unstack()
leg_room_satisfaction_counts.plot(kind='bar', stacked=True)
plt.title('Leg Room Service vs Satisfaction (Stacked Bar Plot)')
plt.xlabel('Leg Room Service Rating')
plt.ylabel('Number of Passengers')
plt.show()
```
